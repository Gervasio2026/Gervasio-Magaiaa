# **Tema:** **Análise de Variância (ANOVA)**

##1.**Introdução e Contextualização Geotécnica**

Na engenharia geotécnica e no estudo de bacias sedimentares, o maior desafio analítico reside na variabilidade natural, anisotropia e heterogeneidade espacial do solo. Como bem cita o João Pedro George: "Se um dia o silêncio tomar-me, amedronta-te, há a variância de mil e um pensamentos." Na mecânica dos solos, a variância também está sempre presente, e a estatística é a ferramenta que nos permite quantificar e conviver com a incerteza dos dados experimentais.

O objetivo desta aula é compreender o fluxo científico e estatístico que valida ensaios laboratoriais e o consequente mapeamento de susceptibilidade a processos erosivos em ambiente de Sistemas de Informação Geográfica (SIG):

**O Processo de Amostragem:** Metodologias para recolher dados representativos de uma região (Amostra) a fim de inferir o comportamento geomecânico de toda a Bacia Sedimentar (População).

**A ANOVA:** Método para testar a igualdade de três ou mais médias populacionais, baseado na análise das variâncias amostrais (conforme a definição clássica da Página 3 do nosso material).

##**2. Desenvolvimento Teórico e Conceitual**

**2.1**
Qual é a diferença entre o Teste **$t$** e o Teste **$F$** (ANOVA)?  normalmente um teste t é usado para examinar as diferenças entre as médias de dois grupos. No entanto, se tivermos mais de dois grupos, não devemos usar apenas vários testes t.

**Por que não?** Porque se realizarmos múltiplos testes t bivariados para comparar os solos dois a dois, o erro se soma (fenómeno conhecido como erro de família). Desse modo, aumenta-se drasticamente a probabilidade de cometer um Erro Tipo I — ou seja, aumenta as chances de encontrar um efeito (afirmar que um solo é diferente do outro) quando, na verdade, não há nenhum.Para contornar o erro de família, quando temos mais de dois grupos para comparar, utilizamos a razão $F$

**2.2** A Razão $F$ e a Partição da VariabilidadeA ANOVA baseia-se na partição da variabilidade total (conforme as Páginas 10 e 14). A razão $F$ mede essencialmente a melhoria devido ao ajuste do modelo (as médias dos grupos versus a média geral de todas as pontuações) e compara isso com o erro restante no modelo (a diferença entre as pontuações reais e as respetivas médias dos grupos).Matematicamente, definimos a estatística de teste como:
$$F = \frac{\text{Variância Sistemática (Entre Grupos)}}{\text{Variância Não Sistemática (Dentro dos Grupos)}}$$

Valores mais altos de $F$ indicam que o efeito do fator testado (ex: a formação geológica) é significativamente maior do que o erro aleatório de medição.

**2.3** Suposições Fundamentais do Modelo

Para que os resultados da ANOVA de um critério (unifatorial) sejam válidos, as amostras devem cumprir os seguintes requisitos:
**Normalidade:** As populações devem ser aproximadamente normalmente distribuídas.
**Homocedasticidade:** As populações devem ter a mesma variância ($\sigma_1^2 = \sigma_2^2 = ... = \sigma_k^2$).
**Independência:** As amostras devem ser aleatórias e mutuamente independentes.

## **3. Exemplo Prático:**
***Coesão de Solos na Bacia do Recôncavo***

Descrição do ProblemaForam realizados ensaios de cisalhamento direto no Laboratório de Geotecnia Ambiental (UFBA) para medir a Coesão do Solo ($c'$ em kPa) — parâmetro geotécnico vital contra a erosão — em 3 formações geológicas distintas da Bacia do Recôncavo: Formação Marizal (A), Formação São Sebastião (B) e Depósitos Quaternários (C). Aplique a ANOVA com nível de significância $\alpha = 0,05$ para verificar se as médias de coesão diferem entre si.

Dados Coletados ($n = 5$ para cada grupo):Formação Marizal (A): $[12, 14, 11, 15, 13]$ kPaFormação São Sebastião (B): $[24, 22, 26, 25, 23]$ kPaDepósitos Quaternários (C): $[17, 19, 18, 21, 20]$ kPaFormulação das Hipóteses (Páginas 7 e 8)Hipótese Nula ($H_0$): $\mu_A = \mu_B = \mu_C$ (As médias de coesão são estatisticamente iguais; o tratamento/fator geológico não tem efeito na resistência).Hipótese Alternativa ($H_1$): Pelo menos uma média é diferente das outras (Não significa que todas as médias sejam diferentes entre si, alguns pares podem ser iguais).


```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols

# 1. ENTRADA DOS DADOS (Coesão c' em kPa)
marizal       = [12, 14, 11, 15, 13]
sao_sebastiao = [24, 22, 26, 25, 23]
quaternario   = [17, 19, 18, 21, 20]

# 2. ESTRUTURAÇÃO DA TABELA (Padrão Tidy Data)
dados = pd.DataFrame({
    'Coesao': marizal + sao_sebastiao + quaternario,
    'Formacao': ['Marizal']*5 + ['Sao_Sebastiao']*5 + ['Quaternario']*5
})

# 3. CÁLCULO DA ANOVA (Baseado no slide compartilhado da UFBA)
# O modelo linear ajusta a Coesão em função da Formação Geológica
modelo = ols('Coesao ~ C(Formacao)', data=dados).fit()
tabela_anova = sm.stats.anova_lm(modelo, typ=1)

# 4. EXIBIÇÃO DOS RESULTADOS NO CONSOLE
print("============================================================")
print("       TABELA ANOVA (DADOS DE COESÃO - RECÔNCAVO)          ")
print("============================================================")
print(tabela_anova)
print("============================================================")

# Extração automática do p-valor e da Estatística F para tomada de decisão
f_calculado = tabela_anova.loc['C(Formacao)', 'F']
p_valor = tabela_anova.loc['C(Formacao)', 'PR(>F)']
alpha = 0.05

print(f"\nRazão F Calculada: {f_calculado:.4f}")
print(f"P-valor obtido: {p_valor:.6f}")

if p_valor < alpha:
    print("\n[VEREDITO]: Rejeitamos H0.")
    print("Conclusão: Existe uma diferença estatisticamente significativa entre as médias (p < 0.05).")
    print("A formação geológica influencia diretamente a coesão do solo contra processos erosivos.")
else:
    print("\n[VEREDITO]: Não rejeitamos H0.")
    print("Conclusão: As médias são estatisticamente iguais. A geologia não influenciou o parâmetro.")

# 5. ILUSTRAÇÃO GRÁFICA (Boxplot + Dispersão dos Pontos Reais)
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

# Gerando o Boxplot didático
sns.boxplot(x='Formacao', y='Coesao', data=dados, width=0.4, palette="Set2", showfliers=False)
# Adicionando os pontos reais coletados no laboratório da UFBA
sns.stripplot(x='Formacao', y='Coesao', data=dados, color='black', size=7, alpha=0.6, jitter=0.05)

plt.title("Variabilidade da Coesão por Formação Geológica\nAnálise de Variância (ANOVA)", fontsize=14, weight='bold')
plt.xlabel("Formação Geológica (Fator)", fontsize=12)
plt.ylabel("Coesão c' (kPa) [Variável Resposta]", fontsize=12)

plt.show()
```

##**4. Conclusão**

 Análise de Variância (ANOVA) fornece o rigor matemático indispensável para validar decisões na engenharia geotécnica e ambiental. Ao decompor a variabilidade total em frações sistemáticas e não sistemáticas através da razão $F$, conseguimos comprovar cientificamente se os horizontes de solo da Bacia do Recôncavo Baiano apresentam comportamentos mecânicos distintos. Esta validação é o passo fundamental para garantir a confiabilidade de zoneamentos de risco de erosão e modelos de estabilidade superficial de encostas executados no SIG.
